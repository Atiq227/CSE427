```python
#np1
import numpy as np

a=np.random.randint(1,101,20)
b=np.random.randint(1,101,20)
print(f"Array a: {a}")
print(f"Array b: {b}")
```

```python
#np2
a = a.reshape(5, 4)
b = b.reshape(5, 4)
print(f"Matrix a:\n{a}\n")
print(f"Matrix b:\n{b}")
```

```python
#np3
scalar = 7
a_res = scalar * a
b_res = scalar * b

print(f"Scalar multiplication of matrix a \n{a_res}\n")
print(f"Scalar multiplication of matrix b \n{b_res}")
```

```python
#np4
b_mod=b.T
c = np.matmul(a, b_mod)
print(f"Matrix c:\n{c}\n")
```

```python
#np5
print(f"2nd and 3rd column of matrix c: \n{c[:, 1:3]}\n")
print(f"Entire matrix using slicing: \n{c[:,:]}")
```

```python
#np6
lar = np.max(c)
index = np.unravel_index(np.argmax(c), c.shape)
print(f"Largest element in matrix c: {lar}")
print(f"Index of the largest element: {index}")
```

```python
#np7
c_mod = c.flatten()
print(f"1D matrix c:\n{c_mod}")
```

```python
import matplotlib.pyplot as plt
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
```

```python
#plt1
plt.figure(figsize=(6,4))
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.xlabel("Temperature in Dhaka")
plt.ylabel("Humidity in Dhaka")
plt.title("Temperature vs Humidity in Dhaka")
plt.show()
```

```python
#plt2
plt.figure(figsize=(12, 8))
plt.plot(study_hours, marks,marker='.', color='orange')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

```python
#plt3
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
sub_marks = [89, 90, 45, 78, 99]

plt.figure(figsize=(10, 6))
plt.barh(subjects, sub_marks, color='black')
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.title("Student's Performance in Different Subjects")
plt.show()
```

```python
#plt4
colors = ['orange', 'green', 'skyblue', 'pink', 'yellow']
plt.figure(figsize=(10, 6))
plt.bar(subjects, sub_marks, color=colors)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Student's Performance in Different Subjects")
plt.show()
```
